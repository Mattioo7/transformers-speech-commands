```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
directory_files = {}

for root, dirs, files in os.walk('./data'):
    directory_files[root] = files
```

```python
counts = {}
for root, files in directory_files.items():
    # Skip the background noise directory
    if root == './data/_background_noise_' or root == './data':
        continue
    
    total_count = len(files)
    unique_count = len(set(file.split("_nohash_")[0] for file in files))
    class_name = os.path.basename(root)
    counts[class_name] = {
        'total_count': total_count,
        'unique_count': unique_count
    }

# Create a DataFrame from the counts dictionary
df = pd.DataFrame.from_dict(counts, orient='index')
df.reset_index(inplace=True)
df.rename(columns={'index': 'class'}, inplace=True)
df.sort_values(by='class', ascending=True, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
total_sum = df['total_count'].sum()
print(f"Total number of files: {total_sum}")
```

```python
# Plot total_count and unique_count for each class
df.plot(x='class', y=['total_count', 'unique_count'], kind='bar', figsize=(15, 7))
plt.title('Total and Unique Counts per Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Count Type')
plt.tight_layout()
plt.show()
```

```python
TARGET_LABELS = ['yes', 'no', 'up', 'down', 'left', 'right', 'on', 'off', 'stop', 'go']
# Create a DataFrame for the target labels
target_df = df[df['class'].isin(TARGET_LABELS)].copy()
unknown_count = df[~df['class'].isin(TARGET_LABELS)][['total_count', 'unique_count']].sum()
unknown_row = pd.DataFrame([{'class': 'unknown', 'total_count': unknown_count['total_count'], 'unique_count': unknown_count['unique_count']}])
target_df = pd.concat([target_df, unknown_row], ignore_index=True)
target_df.sort_values(by='class', ascending=True, inplace=True)
```

```python
target_df
```

```python
target_df.plot(x='class', y=['total_count', 'unique_count'], kind='bar', figsize=(15, 7))
plt.title('Total and Unique Counts per Class (Target Labels)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.legend(['Total Count', 'Unique Count'], title='Count Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
